```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Weightlifting numbers.xlsx')
df
```

```python
df.info()
```

```python
fig,ax = plt.subplots(figsize = (7,5))
ax.scatter(df[df['Jerk Type'] == 'Split']['Front Squat(in Kgs)'] , df[df['Jerk Type']== 'Split']['CnJ'], label="Split",color='darkturquoise')
ax.scatter(df[df['Jerk Type'] == 'Squat']['Front Squat(in Kgs)'] , df[df['Jerk Type']== 'Squat']['CnJ'], label="Squat",color='darkblue')
ax.scatter(df[df['Jerk Type'] == 'Power']['Front Squat(in Kgs)'] , df[df['Jerk Type']== 'Power']['CnJ'], label="Power",color='orange')
ax.set_xlabel('Front Squat (in kg)')
ax.set_ylabel('Clean & Jerk (in kg)')
ax.set_title('Front Squat Vs Cnj')
ax.legend()
plt.show()
```

```python
df.dropna(inplace = True)
```

```python
X = df[['Back Squat (in Kgs)']]
Y = df['Snatch']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
model.predict([[305]])  #Snatch predictor
```

```python
model.intercept_
```

```python
model.coef_
```

```python
line_x = [130,150,200,250,300,360]
line_y = []
for i in line_x:
    y = 0.6047*i -2.691
    line_y.append(y)
line_y
```

```python
fig,ax = plt.subplots(figsize = (6,4))
ax.scatter(X,Y)
ax.plot(line_x, line_y, color = 'lime')
ax.set_xlabel('Back Squat (in kg)')
ax.set_ylabel('Snatch (in kg)')
ax.set_title('Back Squat Vs Snatch')
plt.show()
```

```python
C = df[['Front Squat(in Kgs)']]
D = df['CnJ']
```

```python
mdl = LinearRegression()
```

```python
mdl.fit(C,D)
```

```python
mdl.predict([[115]]) #Cnj Predictor
```

```python
mdl.intercept_
```

```python
mdl.coef_
```

```python
line_a = [110,150,200,250,300]
line_b = []
for i in line_a:
    y = 0.8*i + 9
    line_b.append(y)
line_b
```

```python
fig,ax = plt.subplots(figsize = (6,4))
ax.scatter(C,D)
ax.plot(line_a, line_b, color = 'deepskyblue')
ax.set_xlabel('Front Squat (in kg)')
ax.set_ylabel('Clean & Jerk (in kg)')
ax.set_title('Front Squat Vs Cnj')
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (7,5))
ax.scatter(df[df['Gender'] == 'Male']['Front Squat(in Kgs)'] , df[df['Gender'] == 'Male']['CnJ'], label="CnJ (Male)",color='blueviolet')
ax.scatter(df[df['Gender'] == 'Female']['Front Squat(in Kgs)'] , df[df['Gender'] == 'Female']['CnJ'], label="CnJ (Female)",color='deeppink')
ax.scatter(df[df['Gender'] == 'Male']['Back Squat (in Kgs)'] , df[df['Gender'] == 'Male']['Snatch'], label="Snatch (Male)",color='blue')
ax.scatter(df[df['Gender'] == 'Female']['Back Squat (in Kgs)'] , df[df['Gender'] == 'Female']['Snatch'], label="Snatch (Female)",color='pink')
ax.plot(line_x, line_y, color = 'lime', label = 'Back Squat')
ax.plot(line_a, line_b, color = 'deepskyblue', label = 'Front Squat')
ax.set_title('Male Vs Female Lifting numbers')
ax.set_xlabel('Squat Number (in kg)')
ax.set_ylabel('Snatch / CnJ (in kg)')
ax.legend()
plt.show()
#visualisation of slopes of two diiferent comparisons(i.e. Cnj Vs Front Squat and Snatch Vs Back Squat)
```
